```r
rmvn.eigen <- function (n, mu, Sigma){
    d <- length(mu)
    #F function for eigenvalue decomposition
    ev <- eigen(Sigma, symmetric = TRUE)
    P <- ev$vectors
    Lamba <- ev$values
    Q <- P %*% diag(sqrt(Lamba)) %*% t(P)
    Z <- matrix (rnorm(n*d), nrow=n, ncol=d)
    X <- Z %*% Q + matrix(mu, nrow= n, ncol=d, byrow=TRUE)
    X
}

y <- subset(x=iris, Species=="virginica")[,1:4]
mu <- colMeans(y)
Sigma <-cov(y)
mu
Sigma
X <- rmvn.eigen(1000, mu, Sigma)
#pairs(X)
par(mfrow=c(3,3))
#Sepal.Length - Sepal.Width
plot(X[,1], X[,2], xlab=bquote(x[1]),ylab=bquote(x[2]))
points(y[,1],y[,2],col="red", pch=16)
#Sepal.Length - Petal.Length
plot(X[,1], X[,3], xlab=bquote(x[1]),ylab=bquote(x[3]))
points(y[,1],y[,3],col="red", pch=16)
#Sepal.Length - Petal.Width
plot(X[,1], X[,4], xlab=bquote(x[1]),ylab=bquote(x[4]))
points(y[,1],y[,4],col="red", pch=16)
#Sepal.Width - Petal.Length
plot(X[,2], X[,3], xlab=bquote(x[2]),ylab=bquote(x[3]))
points(y[,2],y[,3],col="red", pch=16)
#Sepal.Width - Petal.Width
plot(X[,2], X[,4], xlab=bquote(x[2]),ylab=bquote(x[4]))
points(y[,2],y[,4],col="red", pch=16)
#Petal.Length - Petal.Width
plot(X[,3], X[,4], xlab=bquote(x[3]),ylab=bquote(x[4]))
points(y[,3],y[,4],col="red", pch=16)
par(mfrow=c(1,1))
```

    test2 <- YourMatrixOfRandomlySampledDirections[,c(1,3)]

```r
runif.sphere <- function(n,d){ #n = number of draws 
    #return a random sample uniformly distributed
    #on the unit sphere in R^d
    M <- matrix(rnorm(n*d), nrow=n, ncol=d)
    L <- apply(M, MARGIN=1, FUN=function(x){sqrt(sum(x*x))})
    D <- diag(1/L)
    U <- D %*% M 
    U
}
#generate a sample in d=3 and plot
X <- runif.sphere(200,3)

par(mfrow=c(2,2))
test1 <- X[,1:2]
par(pty="s")
plot(test1, xlab=bquote(x[1]),ylab=bquote(x[2]))
par(pty="m")

test2 <- X[,c(1,3)]
par(pty="s")
plot(test2, xlab=bquote(x[1]),ylab=bquote(x[3]))
par(pty="m")

test3 <- X[,2:3]
par(pty="s")
plot(test3, xlab=bquote(x[2]),ylab=bquote(x[3]))
par(pty="m")
par(mfrow=c(1,1))
```

#### a) Similarly to HW2, using simulation, please approximate the p-value that we would see data as extreme as this data (one sided test using the same distance statistic as before) if the students really were sitting randomly, independently uniformly around the classroom. Make sure you justify which direction you test in.

#### b) Then answer: what is your conclusion about how students sat in class? Use hypothesis testing at the $\alpha=.05$ level.


#### 2a)
p_value = 0.033

#### 2b)
$H_0$ : Students are sittting uniformly in the classroom.<br>
$H_a$ : Students are not sitting uniformly in classroom.

$p < 0.05$ Therefore we reject $H_0$, and conculde that there is sufficient evidence to say students are not sitting uniformly in classroom.


```r
#ACTUAL DATA
#position of students
x_vec = c(.1, .2, .4, .5, .6)
y_vec = c(.8, .9, .3, .5 ,.7)

#calculates distance between students
distance_stat = function(x_vec, y_vec){
    
    total_distance = 0 # initialize
    
    for(i in 1:(length(x_vec)-1)){
        
        for(j in (i+1):length(x_vec)){
            
            pair_distance = sqrt((x_vec[i]-x_vec[j])^2 + (y_vec[i]-y_vec[j])^2)
            total_distance = total_distance + pair_distance
        }
    }
    
    return(total_distance)
}
s = distance_stat(x_vec, y_vec)
s
```

```r
#SIMULATION
#class_distance creates uniform r.v's with x coordinates [0,2], y coordinates [0,1]
#Since width of classroom is 2 unit, draws uniform r.v's with x coordinates [0,2]
#Since lenght of classroom is 1 unit, draws uniform r.v's with y coordinates [0,1]
#The shape of classroom is trapezoid. If right wall of classroom is a function, 
#it will be y = (10/7) x - (13/7) 
#if y > (10/7) x - (13/7) which means the r.v. is locates left side of right wall
#then it will accept the r.v.
#otherwise, if x coordinate of r.v. is less than 1.3, than it will accept the r.v.

class_distance <- function(n_student){
    k <- 0
    x <- vector(length=n_student)
    y <- vector(length=n_student)
    save_dist <- vector(length=n_student)
    while(k<n_student){
        unif_draw_x = runif(1, min=0, max=2) 
        unif_draw_y = runif(1, min=0, max=1)
        if (unif_draw_x >= 1.3 && unif_draw_y >=(((10/7)*unif_draw_x)-(13/7))){
            k=k+1
            x[k]=unif_draw_x
            y[k]=unif_draw_y
        } else if (unif_draw_x <= 1.3){
            k=k+1
            x[k]=unif_draw_x
            y[k]=unif_draw_y
        }
        
    }
    unif_f=distance_stat(x,y)
    return(unif_f)
}
```

```r
#Monte Carlo Simulation
#This part is not necessary for answer (because I will calculate distances next code)
#I just want to see mean of distances between simulated coordination.
n_draws = 1000
n_student=5
save_distances = vector(length=n_draws)
for(i in 1:n_draws){
    unif_s = class_distance(n_student)
    save_distances[i] = unif_s
}
mean(save_distances)
```

```r
#Continue the Monte Carlo algorithm

n_student = 5
n_draws = 1000
indicator_vec = vector(length=n_draws) # to save indicators
for(i in 1:n_draws){
    unif_s = class_distance(n_student)
    indicator = as.integer(unif_s < s) #checks simulated distance is less than actual data d.
    indicator_vec[i] = indicator
}

mean(indicator_vec)
```

Specifically, check your code by checking that your function standardizes these three sets of $\overrightarrow{X}$:

$\overrightarrow{X_1} \sim N_2(\overrightarrow{\mu}_1, \Sigma_1)$ where $\overrightarrow{\mu}_1$ = (5, -3) and  $\Sigma_1 = \left[{\begin{array}{cc}

$\overrightarrow{X_3} \sim N_4(\overrightarrow{\mu}_3, \Sigma_3)$ where $\overrightarrow{\mu}$ = (1, 0, -1, 0) and  $\Sigma_3 = \left[{\begin{array}{cccc}
 1 & 0.3 & 0.09  & 0 \\

0.09 & 0.3 & 1 & 0.3 \\
 0 & 0.09 & 0.3 & 1
\end{array}}\right]$

### Solution:
To go from $\mathbb{X}$ to $\mathbb{Z}$, the procedure is inverted:
1. Generate $\overrightarrow{X} \sim N_d(\overrightarrow{\mu}, \Sigma)$

2. Find $\Sigma^{-1/2}$. I used Choleski decomposition but, the eigenvalue decomposition alsa can be used.

3. Apply standardize transformation. $\mathbb{Z} = (\mathbb{X}-\mathbb{J}\overrightarrow{\mu})\Sigma^{-1/2}$

4. Print $\mathbb{Z}$ and check the mean and variance. $\mathbb{Z}$ should be $\overrightarrow{Z} \sim N_d(0,1)$)

```r
library (MASS)
n <- 100
#creates X multivariate normal distribution with mean mu, variance sigma
#dimention of X is n x d
generate.X <- function(n, mu, sigma){
    X <- mvrnorm(n, mu, sigma)
    return(X)
}

std.rmvn <- function(k){
  n <- nrow(k) 
  d <- ncol(k)
  mu <- apply(k, MARGIN=2, mean)
  Sigma <- cov(k)
  s <- chol(solve(Sigma)) #sigma^-1/2
  Z <- (apply(k, MARGIN= 1,FUN=function(x){ s %*% matrix(x - mu)} ))
  return(t(Z))
}
```

```r
mu1 <- c(5, -3)
sigma1 <- matrix(c(1, -0.9, -0.9, 1), nrow = 2, ncol = 2)
A <- generate.X(n, mu1, sigma1)
X <- std.rmvn(A)

## Means and covariance 
#After standardize a multivariate normal sample, I expectto see the mean vector = (0,0) sigma=identity matrix
#For X_1 with mean mu_1 and variance sigma1
Mean1 <- round(apply(X, 2, mean))
Sigma1 <- round(cov(X))
Mean1
Sigma1
```
